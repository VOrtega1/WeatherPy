#### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
locations = CityData_df[['Lat', 'Lng']]
locations
```

```python
hum_weight = CityData_df[['Humidity']]
hum_weight
```

```python
# Store latitude and longitude in locations
locations = CityData_df[["Lat", "Lng"]]

# Fill NaN values and convert to float

hum_weight = CityData_df["Humidity"].fillna(0)
hum_weight = CityData_df["Humidity"].astype(float)
```

```python
# Plot Heatmap
fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=hum_weight, 
                                 dissipating=False, max_intensity=10,
                                 point_radius=1)


# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```
